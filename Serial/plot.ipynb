```python
plt.figure(figsize=(10, 6))
for N, speed_ups in speed_up_data.items():
    plt.plot(num_threads, speed_ups, marker='o', label=f'N={N}')
plt.xlabel('Number of Processing Elements (p)')
plt.ylabel('Speed Up')
plt.title('(Population Generation) Speed Up vs. Number of Processing Elements for Different N')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for i, p in enumerate(num_threads):
    speeds = [speed_up_data[N][i] for N in population_sizes]
    plt.plot(population_sizes, speeds, marker='o', label=f'p={p}')
plt.xlabel('Population Size (N)')
plt.ylabel('Speed Up')
plt.title('(Population Generation) Speed Up vs. Population Size for Different Processing Elements')
plt.xscale('log')  # Since the range of N is large, a logarithmic scale may be clearer
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Speedup data for each function
speed_up_data_pop = {
    100000: [0.189718, 0.375303, 0.61753, 0.901163, 1.03333, 1.04027],
    500000: [0.183414, 0.345803, 0.64032, 0.879268, 0.989026, 1.02125],
    1000000: [0.172333, 0.333015, 0.612747, 0.883957, 0.966043, 0.962043],
    5000000: [0.195263, 0.306688, 0.680566, 0.877692, 1.07138, 1.08367],
    10000000:  [0.189941, 0.360782, 0.648986, 0.936779, 1.01778, 0.957595]
}

speed_up_data_fit = {
    100000: [0.2, 0.364035, 0.628788, 0.965116, 0.697479, 0.592856],
    500000: [0.365088, 0.626484, 1.01932, 1.62945, 1.23604, 1.03782],
    1000000: [0.206284, 0.656205, 1.047, 1.62172, 1.20586, 0.985497],
    5000000: [0.1464, 0.685504, 1.07317, 1.60018, 1.20374, 0.889726],
    10000000: [0.415731, 0.710686, 1.10746, 1.62807, 1.20215, 0.978028]
}

speed_up_data_cross = {
    100000: [0.978365, 1.03846, 0.989177, 0.99043, 1.05814, 1.01786],
    500000: [1.07732, 1.02134, 1.00317, 0.99115, 0.985437, 0.985545],
    1000000: [0.976211, 1.00467, 0.969017, 0.982547, 1.032, 0.922297],
    5000000: [1.01888, 1.03302, 1.01065, 1.00753, 1.01935, 0.999357],
    10000000: [1.00214, 1.00889, 1.01007, 1.01302, 1.00611, 1.00836]
}

# Number of processing units (p)
num_threads = [1, 2, 4, 8, 16, 32]

# Function to calculate parallel efficiency
def calculate_parallel_efficiency(speed_up_data, num_threads):
    parallel_efficiency = {}
    for N, speed_ups in speed_up_data.items():
        parallel_efficiency[N] = [speed_up / p * 100 for speed_up, p in zip(speed_ups, num_threads)]
    return parallel_efficiency

# Calculate parallel efficiency for each function
parallel_efficiency_pop = calculate_parallel_efficiency(speed_up_data_pop, num_threads)
parallel_efficiency_fit = calculate_parallel_efficiency(speed_up_data_fit, num_threads)
parallel_efficiency_cross = calculate_parallel_efficiency(speed_up_data_cross, num_threads)

# Plot parallel efficiency for each function
def plot_parallel_efficiency(parallel_efficiency, title):
    for N, efficiencies in parallel_efficiency.items():
        plt.plot(num_threads, efficiencies, label=f"N={N}")
    plt.title(title)
    plt.xlabel("Number of Processing Units (p)")
    plt.ylabel("Parallel Efficiency (%)")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot parallel efficiency for each function
plot_parallel_efficiency(parallel_efficiency_pop, "Parallel Efficiency for Population Initialization")
plot_parallel_efficiency(parallel_efficiency_fit, "Parallel Efficiency for Fitness Calculation")
plot_parallel_efficiency(parallel_efficiency_cross, "Parallel Efficiency for Crossover and Mutation")
```

Serial vs Parallel

```python
import matplotlib.pyplot as plt

population_sizes = [100000, 500000, 1000000, 5000000, 10000000]
total_durations = [2.775, 13.643, 25.534, 122.92, 229.685]
total_durations1 = [0.363, 1.768, 3.486, 18.53, 47.174]

plt.figure(figsize=(8, 6))

# Plotting serial total durations
plt.plot(population_sizes, total_durations, marker='o', color='b', linestyle='-', label='Serial')

# Plotting parallel total durations
plt.plot(population_sizes, total_durations1, marker='o', color='r', linestyle='-', label='Parallel')

plt.title('Total Duration vs. Population Size')
plt.xlabel('Population Size')
plt.ylabel('Total Duration (seconds)')
plt.grid(True)
plt.xticks(population_sizes)
plt.tight_layout()

# Add annotations for each point in serial plot
for i in range(len(population_sizes)):
    plt.text(population_sizes[i], total_durations[i], f"{total_durations[i]:.3f}", ha='right', va='bottom', color='b')

# Add annotations for each point in parallel plot
for i in range(len(population_sizes)):
    plt.text(population_sizes[i], total_durations1[i], f"{total_durations1[i]:.3f}", ha='right', va='bottom', color='r')

# Add legend
plt.legend()

plt.show()
```
